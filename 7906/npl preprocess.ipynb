```python
import numpy as np
from sklearn.feature_extraction.text \
    import CountVectorizer
count = CountVectorizer()
docs = np.array(['The sun is shining',
                'The weather is sweet',
                'The sun is shining, the weather is sweet,'
                'and one and one is two'])
count.fit(docs)
```

```python
print(count.vocabulary_)
```

```python
bag = count.transform(docs).toarray()
```

```python
from sklearn.feature_extraction.text \
    import TfidfTransformer

tfidf = TfidfTransformer(use_idf=True,
                        norm='l2',
                        smooth_idf=True)
np.set_printoptions(precision=2)
tfidf.fit(bag)
```

```python
tfidf.transform(bag).toarray()
```

```python
bag1 = count.transform(["shining one two and two"]).toarray()
bag1
```

```python
tfidf.transform(bag1).toarray()
```

```python
def tokenizer(text):
    return text.split()

words = 'runners like running and is a the thus they run shining'
```

```python
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()

def tokenizer_porter(text):
    return [porter.stem(word) for word in
          text.split()]
tokenizer_porter(words)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
[w for w in tokenizer_porter(words) if w not in stop]
```
